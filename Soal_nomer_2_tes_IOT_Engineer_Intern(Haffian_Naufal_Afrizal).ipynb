````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
#klasifikan data dari tabel sesuai kolom
voltage_data = np.array([
    2951.6, 2936.5, 2815.6, 2486.8, 2329.6, 2015.4,
    1732.2, 1567, 1311.8, 1216.4, 1065.8, 859.4,
    680.75, 548.8, 490.2, 344.6, 262, 129, 47.2, 43 ])

volume_data = np.array([
    13.5, 18, 22.5, 27, 31.5, 36,
    40.5, 45, 49.5, 54, 58.5, 63,
    67.5, 72, 76.5, 81, 85.5, 90, 94.5, 99  ])

# Function to get fuel volume based on sensor voltage
def get_fuel(tegangan_sensor: float) -> float:
    # Interpolate to find the corresponding volume for the given voltage
    volume = np.interp(tegangan_sensor, voltage_data, volume_data)
    return volume

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(voltage_data, volume_data, marker='o', linestyle='-', color='b')
plt.title('Voltage vs Fuel Volume')
plt.xlabel('Voltage (mV)')
plt.ylabel('Volume (L)')
plt.grid()
plt.xlim(0, 3000)
plt.ylim(0, 100)
plt.show()
````
